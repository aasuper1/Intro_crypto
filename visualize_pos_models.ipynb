```python
import pandas as pd
import quandl
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate

plt.rcParams['figure.figsize'] = [20, 20]
plt.rcParams.update({'font.size': 15})
```

```python
pPos_res = pd.read_csv("pPOS_results.csv")
dPos_res = pd.read_csv("dPOS_results.csv")
ePos_res = pd.read_csv("ePOS_results.csv")
lottery_res = pd.read_csv("lottery_results.csv")
lottery_res_l = pd.read_csv("lottery_results_low_cost.csv")
lottery_res_h = pd.read_csv("lottery_results_high_cost.csv")

pPos_res_r = pd.read_csv("pPOS_real_results.csv")
dPos_res_r = pd.read_csv("dPOS_real_results.csv")
ePos_res_r = pd.read_csv("ePOS_real_results.csv")
lottery_res_r = pd.read_csv("lottery_real_results.csv")

x_p,y_p=pPos_res["iteration"], pPos_res["gini coefficient"]
x_pr,y_pr=pPos_res_r["iteration"], pPos_res_r["gini coefficient"]
plt.plot(x_p,y_p, '--bo', label='Pure Proof of Stake')

x_d,y_d=dPos_res["iteration"], dPos_res["gini coefficient"]
x_dr,y_dr=dPos_res_r["iteration"], dPos_res_r["gini coefficient"]
plt.plot(x_d,y_d, '-go', label='Delegated Proof of Stake')

x_e,y_e= ePos_res["iteration"], ePos_res["gini coefficient"]
x_er,y_er= ePos_res_r["iteration"], ePos_res_r["gini coefficient"]
plt.plot(x_e,y_e, '--mo', label = 'Extended Proof of Stake')

x_l,y_l= lottery_res["iteration"], lottery_res["gini coefficient"]
x_lr,y_lr= lottery_res_r["iteration"], lottery_res_r["gini coefficient"]
x_ll,y_ll= lottery_res_l["iteration"], lottery_res_l["gini coefficient"]
x_lh,y_lh= lottery_res_h["iteration"], lottery_res_h["gini coefficient"]
plt.plot(x_l,y_l, '--yo',label = 'Raffle Proof of Stake' )

plt.xlabel("Iteration Number")
plt.ylabel("Gini Coefficient")
plt.title("Gini Coefficient vs Iteration Number")

plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x_p, y_p, label='Pure Proof of Stake')
axs[0, 0].plot(x_d, y_d, 'tab:orange', label='Delegated Proof of Stake')
axs[0, 0].plot(x_e, y_e, 'tab:green', label='Extended Proof of Stake')
axs[0, 0].plot(x_l, y_l, 'tab:red', label='Raffle Proof of Stake')
axs[0, 0].set_title('Normal Distribution')
axs[0, 0].legend()
axs[0, 1].plot(x_pr, y_pr, label='Pure Proof of Stake')
axs[0, 1].plot(x_dr, y_dr, 'tab:orange', label='Delegated Proof of Stake')
axs[0, 1].plot(x_er, y_er, 'tab:green', label='Extended Proof of Stake')
axs[0, 1].plot(x_lr, y_lr, 'tab:red', label='Raffle Proof of Stake')
axs[0, 1].set_title('Real Distribution')
axs[0, 1].legend()
axs[1, 0].plot(x_ll, y_ll, 'tab:green', label='Ticket Cost: 0.1 Coin')
axs[1, 0].plot(x_lh, y_lh, 'tab:red', label='Ticket Cost: 5 Coin')
axs[1, 0].set_title('Effects of Ticket Costs')
axs[1, 0].legend()
for ax in axs.flat:
    ax.set(xlabel='Iteration Number', ylabel='Gini Index')

# Hide x labels and tick labels for top plots and y ticks for right plots.
# for ax in axs.flat:
#     ax.label_outer()
```
